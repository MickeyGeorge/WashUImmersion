```python
import pandas as pd
import numpy as np
import xgboost as xgb

train = pd.read_csv("./all/train.csv")
test = pd.read_csv("./all/test.csv")
train.head()
```

```python
# Feature engineering
# Compute the following features:
#   - followers/following
#   - mentions received/sent
#   - retweets received/sent
#   - followers/posts
#   - retweets received/posts
#   - mentions received/posts

def ratio(df, user, name1, name2, new_name, factor = 1):
    df[user + "_" + new_name] = (df[user + "_" + name1]+factor)/(df[user + "_" + name2]+factor)
    return df

def get_data(df):
    new_data_set = df.iloc[:, 1:23]
    for user in ["A", "B"]:
        new_data_set = ratio(new_data_set, user, "follower_count", "following_count", "follower_ratio")
        new_data_set = ratio(new_data_set, user, "mentions_received", "mentions_sent", "mention_ratio")
        new_data_set = ratio(new_data_set, user, "retweets_received", "retweets_sent", "retweet_ratio")
        new_data_set = ratio(new_data_set, user, "follower_count", "posts", "follow_post")
        new_data_set = ratio(new_data_set, user, "retweets_received", "posts", "retweet_post")
        new_data_set = ratio(new_data_set, user, "mentions_received", "posts", "mention_post")
    names = ["_".join(x.split("_")[1:]) for x in [x for x in new_data_set.columns if "A" in x]]
    for name in names:
        new_data_set[name+"_diff"] = new_data_set["A_"+name] - new_data_set["B_"+name]
    X = new_data_set
    y = df.iloc[:,0]
    data_dmatrix = xgb.DMatrix(data=X,label=y)
    return data_dmatrix
```

```python
# Use cross-validation to find the best parameter
data_dmatrix = get_data(train)
params = {"objective":"binary:logistic",'eval_metric': 'auc', 
          "eta":0.005, "gamma":1, "lambda":3, "max_depth":4, "min_child_weight":1,
          "verbose":False, "subsample":0.8, "colsample_bytree":0.8}

cv_result = xgb.cv(dtrain = data_dmatrix, nfold = 5, params=params,num_boost_round=3000)
np.max(cv_result["test-auc-mean"])
```

```python
# Train the model
model = xgb.train(dtrain = data_dmatrix, params=params,num_boost_round=3000)
```

```python
# Prepare the testing data and predict on it
def process_test_data(df_test):
    new_data_set = df_test
    for user in ["A", "B"]:
        new_data_set = ratio(new_data_set, user, "follower_count", "following_count", "follower_ratio")
        new_data_set = ratio(new_data_set, user, "mentions_received", "mentions_sent", "mention_ratio")
        new_data_set = ratio(new_data_set, user, "retweets_received", "retweets_sent", "retweet_ratio")
        new_data_set = ratio(new_data_set, user, "follower_count", "posts", "follow_post")
        new_data_set = ratio(new_data_set, user, "retweets_received", "posts", "retweet_post")
        new_data_set = ratio(new_data_set, user, "mentions_received", "posts", "mention_post")
    names = ["_".join(x.split("_")[1:]) for x in [x for x in new_data_set.columns if "A" in x]]
    for name in names:
        new_data_set[name+"_diff"] = new_data_set["A_"+name] - new_data_set["B_"+name]
    return new_data_set

df_test = process_test_data(test)
preds = model.predict(xgb.DMatrix(df_test), ntree_limit=model.best_ntree_limit)
```

```python
df_submission = pd.DataFrame({"Id":range(1,len(preds)+1), "Choice":preds})
df_submission.to_csv("~/Downloads/submission.csv", index=False)
```
